# 离散傅里叶变换

#### 产生原始时域采样序列

## 定义
离散频率序列 $X(m)$ 定义如下：  
$$ X(m) = \sum_{n=0}^{N-1}x(n)e^{-j2\pi nm/N}$$  
$N$ 是时域采样点点数  
$m$ 是频域频率轴上的点

## DFT对称性
当输入序列 $x(n)$ 是实数时，$X(m)$ 是 $X(N-m)$ 的复共轭，即：
$$ X(m)=X^*(N-m)$$  
推导  
$$ \begin{aligned}
X(N-m)&=\sum_{n=0}^{N-1}x(n)e^{-j2\pi n(N-m)/N} \\
& =\sum_{n=0}^{N-1}x(n)e^{-j2\pi nN/N}e^{-2j\pi n(-m)/N} \\
& =\sum_{n=0}^{N-1}e^{-j2\pi n}e^{j2\pi nm/N} \\
\end{aligned} $$
由对所有的整数值 $n$ 有 $e^{-j2\pi n}=cons(2\pi n)-jsin(2\pi n)=1$，所以  
$$ X(N-m)=\sum_{n=0}^{N-1}x(n)e^{j2\pi nm/N}$$

## DFT 时域移位定理
如果决定从$n$等于某个整数$k$开始对$x(n)$采样，与$n=0$形成对比，令该时移样值的DFT是$X_{shifted}(m)$，则  
  
  $$ X_{shifted}(m)=e^{j2\pi km/N}X(m)$$  
  
$x(n)$ 的起始采样向右移$k$个位置，那么DFT输出$X_{shifted}(m)$的频谱就是把每个$X(m)$复数项乘以$e^{j2\pi km/N}$，这仅仅是$2\pi km/N$弧度或$360km/N$度的相移。  

如果$x(n)$ 的起始采样向左移$k$个位置，那么DFT输出$X_{shifted}(m)$的频谱就是把每个$X(m)$复数项乘以$e^{-j2\pi km/N}$  
这也意味着，__时域的时移对应着频域的相移__。时域上的固定时移，对应着频域上不同频点的不同相移。频率越高，相移越大。一个固定的时移，对应的相移与周期成反比，也就是与频率成正比。

### 例1：DFT时移
对于原始输入函数：  
$$ x_m(t)=sin(2\pi 1000t)+0.5sin(2\pi 2000t+3\pi /4) $$  
  
如果以$t_s=1/f_s$秒，采样频率为$f_s=8000$样值/秒 对函数进行采样。设采样点数为$N=8$，则DFT会对8个点进行运算。所以，8个元素的序列$x(n)$等价于在$nt_s$时对$x_{in}(t)$的瞬时样值：  
$$ x(n)=x_{in}(nt_s)=sin(2\cdot \pi \cdot 1000\cdot n\cdot t_s)+0.5\cdot sin(2\cdot \pi \cdot 2000\cdot n\cdot t_s+3\cdot \pi /4) $$  

对输入序列延迟$k=3$样值进行采样。延迟后产生的新的序列$x(n)$序列为:  
$$ [ 1.0607\quad0.3535\quad-1.0607\quad-1.3535\quad-0.3535\quad0.3535
  \quad0.3535  \quad0.6464] $$  
  
对这序列进行DFT变换，可以求得变换后的幅频特性和相频特性图。图中在 $m=1$ 的地方有分量，并且可以求得 $X_{shifted}(1)$ 的幅度-相位形式：  
$$ X_{shifted}(1)=e^{j2\pi km/N}\cdot X(1)=e^{j2\pi 3\cdot 1/8}\cdot 4e^{-j\pi /2}=4e^{j(6\pi /8-4\pi /8)}=4e^{j\pi /4}$$
  
所以，$X_{shifted(1)}$ 的幅度为4，相角为$\pi /4$或者$+45^{\circ}$   

```python
from numpy import *
from matplotlib import pyplot as plt
def funSignal(n, fs=8000.0):    
    ts = 1/fs
    a = sin(2*pi*1000*n*ts)
    b = 0.5*sin(2*pi*2000*n*ts+pi*3/4)
    return  a+b

n = array([0,1,2,3,4,5,6,7],dtype=float);newfs = 8000.0;
Xin0 = funSignal(n, newfs);
plt.figure(figsize=(10,4));
plt.title(r'$x[n] = sin(2\pi 1000nt_{s})+sin(2\pi 2000nt_{s}+3\pi /4)$, \
          $n = [0,1,2,3,4,5,6,7],  fs=8kHz$',fontsize=10);
plt.plot(funSignal(linspace(0,10,1000), newfs),alpha=1);plt.stem(n*100,Xin0,'--');
out0 = around(fft.fft(Xin0),4);
plt.figure(figsize=(10,4));
plt.title('Amp0',fontsize=10);plt.stem(n,abs(out0),'--');
plt.figure(figsize=(10,4));
plt.title('Phase0',fontsize=10);plt.stem(n,angle(out0)*180/pi,'--');
plt.show()
print 'Out0 Amplitude:',abs(out0),'\n'
print 'Out0 phase:',angle(out0)*180/pi
```

利用频率轴计算公式，得知$f(1)=\frac{1\cdot 8000}{8}=1kHz$。
在相频图上可以看到在$1kHz$的频点上，有-90度的信号。
同理，在$2kHz$上面，有45度的信号。

#### 产生经过时移的采样序列

```python
Xin1 = funSignal(n+3, newfs); 
plt.figure(figsize=(10,4));
plt.title(r'$x[n] = sin(2\pi 1000nt_{s})+sin(2\pi 2000nt_{s}+3\pi /4)$, \
          $n = [3,4,5,6,7,8,9,10],  fs=8kHz$',fontsize=10);
plt.plot(funSignal(linspace(0,10,1000), newfs),alpha=1);plt.stem((n+3)*100,Xin1,'--')
out1 = around(fft.fft(Xin1),4);
plt.figure(figsize=(10,4));
plt.title('Amp1',fontsize=10);plt.stem(n,abs(out1),'--');
plt.figure(figsize=(10,4));
plt.title('Phase1',fontsize=10);plt.stem(n,angle(out1)*180/pi,'--');
plt.show()
```

采样序列经过时移后，可以看到做DFT后的频谱图跟预期一样仍然没有改变。但是相位图却发生了变换，所以频域相位图能反映出时域的时延。  
可以计算出，当前$1kHz$上的信号，相位是$\pi /4$
$$ X_{shifted}(1)=e^{j2\pi km/N}\cdot X(1)=e^{j2\pi 3\cdot 1/8}\cdot 4e^{-j\pi /2}=4e^{j(6\pi /8-4\pi /8)}=4e^{j\pi /4}$$  
所以，该相位是初始相位与时移后叠加的相位。那么时移3个采样点，对$1kHz$产生的相位偏移就是$e^{j2\pi 3\cdot 1/8}=e^{j3\pi /4}$的相位$3\pi /4$。

#### 把原始序列在频域进行相位处理，与前面时移做比较

### 例2：频偏估计

## DFT频率轴
对于该DFT生成的频谱图，应当能准确计算出频率轴上坐标点的意义。如果计算$m=1$处的频率，可以由以下式子算出：  
$$ f_{analysis}(m)=\frac{mf_s}{N}=f_{analysis}(1)=\frac{1\cdot 8000}{8}=1kHz$$
并且，DFT频域上频率间距(分辨率)是 $f_s/N$

## 卷积性质
若$x[n]$，$h[n]$和$y[n]$分别为某一线性时不变系统的输入、单位脉冲响应和输出，而有
$$y[n]=x[n]*h[n]$$
那么
$$Y(e^{j\omega})=X(e^{j\omega})H(e^{j\omega})$$

### 例1：线性时不变系统的模-相特性
考虑一个线性时不变系统，其单位脉冲响应为
$$h[n]=\delta[n-n_0]$$
它的频率响应$H(e^{j\omega})$就是
$$H(e^{j\omega})=\sum_{n=-\infty}^{+\infty}\delta[n-n_0]e^{-j\omega n}=e^{-j\omega n_0}$$
因此，对于傅里叶变换为$X(e^{j\omega})$的任意输入$x[n]$，其输出的傅里叶变换是
$$Y(e^{j\omega})=e^{-j\omega n_0}X(e^{j\omega})$$
于是这个输出与傅里叶的时移定理是一致。同时$H(e^{j\omega})=e^{-j\omega}$，意味着是一个纯时移系统，对所有频率其模贡献是1x，而相移则与频率成线性关系，即$-n_0\omega$，斜率是$-n_0$，也即是时移的值。

## DFT频谱泄露
假设以#f_s#采样率对输入进行N个点的采样，并做DFT。那么DFT会产生N个点的变换，并且其输出对应的频率与$f_s$、$N$相关：
$$f_{analysis}(m)=\frac{mf_s}{N}$$
其中，$m=0,1,2,...,N-1$  
于是，只有当输入的信号刚好落在离散的频点$f_{analysis}(m)$上，其DFT输出才能有正确的结果。  
  
先引入一个性质，对于$k$个周期上，有$N$点输入时间序列的实余弦输入信号。从DFT输出的频率轴索引$m$的角度来说，$N$点DFT的幅频响应由下面的sinc函数近似：
$$X(m)=\frac{A_0N}{2}\cdot \frac{sin[\pi(k-m)]}{\pi(k-m)}$$
其中$A_0$是DFT输入正弦曲线的峰值。
  
  下面的例子，是以$f_s=32000Hz$采样率进行采样，取出32点做DFT。分别对信号$8kHz$和$8.5kHz$进行处理。从8.5kHz的图中可以看出，非整数倍的周期数采样，会导致所有DFT输出的频率点上都具有非零的幅度。因为，以$f_s=32000$采样，取32个点，相当于时长$\frac{1}{32000}\cdot 32=\frac{1}{1000}s$。当输入信号是$8kHz$时，$\frac{1}{1000}/\frac{1}{8000}=8$刚好是整数倍的周期数，而$8.5kHz$信号的周期数是$\frac{1}{1000}/\frac{1}{8500}$不是整数倍。于是造成了频率泄露。

```python
def funSignaln(n, frequency=8000.0,fs=32000.0):    
    ts = 1/fs
    a = sin(2*pi*frequency*n*ts)
    return  a
srcn=linspace(-16,16,32000,endpoint=False,dtype=float);
freSrc = 32/2*sinc(srcn);
n0=linspace(0,31,32);
fre0 = funSignaln(n0,frequency=8000.0,fs=32000.0);
fre0out = around(fft.fft(fre0),4);
plt.figure(figsize=(10,4));
plt.plot(srcn[11000:21000],abs(freSrc)[11000:21000],lw=0.5,alpha=0.5);
plt.title('Fre0 Amp',fontsize=10);plt.stem(linspace(-5,5,11),abs(fre0out)[3:14],'--');
ax=plt.gca();ax.set_xticklabels([3,4,6,8,10,12])
plt.show()
```

## 加窗
为了避免因为非整周期的采样导致DFT后的频率泄露，可以在做采样后进行加窗处理。先给出定义，假设原始$N$个输入信号样值由$n$来索引，其中$0\leqslant n\leqslant N-1$，将$N$个时域的窗系数称为$w(n)$;也就是说一个输入序列$x(n)$在进行DFT之前先乘以相应的窗系数$w(n)$。所以窗口化的输入信号$x(n)$的DFT，$X_w(m)$，有以下形式：
$$X_w(m)=\sum_{n=0}^{N-1}w(n)\cdot x(n)e^{-j2\pi nm/N}$$

### 矩形窗
$$w(n)=1,\,\,\, n=0,1,2,...,N-1$$

### 三角窗
$$w=
\begin{cases}
\frac{n}{N/2}, & n=0,1,2,...,N-1 \\
2-\frac{n}{N/2}, & n = N/2+1,N/2+2,...,N-1 \\
\end{cases}
$$

### Hanning窗(升余弦窗，Hann窗或 von Hann窗)
$$w(n)=0.5-0.5cos(\frac{2\pi n}{N}), \,\,\, n =0,1,2,...,N-1$$

### Hamming窗
$$w(n)=0.5-0.46cos(\frac{2\pi n}{N}), \,\,\, n =0,1,2,...,N-1$$

## 离散时间傅里叶变换性质
$$X_k=\sum_{n=-\infty}^{+\infty}x[n]e^{-jk(2\pi/N)n}$$
令$\omega=e^{jk(2\pi/N)}$，则：
$$X(e^{j\omega})=\sum_{n=-\infty}^{+\infty}x[n]e^{-j\omega n}$$
$$x[n]=\frac{1}{2\pi}\int_{2\pi}X(e^{j\omega})e^{j\omega n}d\omega$$

### 离散时间傅里叶变换的周期性
离散时间傅里叶变换对于$\omega$ 来说总是周期的，其周期为$2\pi$，即：
$$ X(e^{j(\omega +2\pi)})=X(e^{j\omega})$$  
  
  
  

### 线性性质
若有  
$$ x_1[n]\longleftrightarrow X_1(e^{j\omega})$$
且有
$$ x_2[n]\longleftrightarrow X_2(e^{j\omega})$$
则
$$ ax_1[n]+bx_2[n]
\longleftrightarrow 
aX_1(e^{j\omega})+bX_2(e^{j\omega})$$

### 时移与频移性质
若
$$ x[n]\longleftrightarrow X(e^{j\omega})$$
则有时移性质：
$$ x[n-n_0]\longleftrightarrow e^{-j\omega n_0}X(e^{j\omega})$$
和频移性质：
$$ e^{j\omega _0n}x[n]\longleftrightarrow X(e^{j(\omega -\omega _0)})$$
